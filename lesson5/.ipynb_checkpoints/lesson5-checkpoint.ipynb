**Задачи для 5 го занятия которые мы разбирали на занятии**

<img src="9.187.png">

```python
# s = input("Введите натуральное число")

nums = ["ноль", "один", "два", "три", "четыре", "пять", "шесть", "семь", "восемь", "девять"]

def get_num_as_str(s):
    '''
        Завернул задачу в функцию 
        чтобы было легче проверить
    '''
    if s == "1000":
        return "тысяча"
    
    res = ""

    # сотки
    if len(s) >= 3:
        hundreds = int(s[-3])
        if hundreds == 1:
            res += "сто "
        elif hundreds == 2:
            res += "двесте "
        elif hundreds in [3,4]:
            res += nums[hundreds] + "ста "
        else:
            res += nums[hundreds] + "сот "


    # десятки
    voy = False
    if len(s) >= 2:
        tens = int(s[-2])
        if tens == 1:
            # тут нужно поковыряться
            # на следующем шаге
            voy = True
        elif tens in [2, 3]:
            res += nums[tens] + "дцать "
        elif tens == 4:
            res += "сорок "
        elif tens in [5, 6, 7, 8]:
            res += nums[tens] + "десят "
        else:
            res += "девяносто "
    
    # единицы
    units = int(s[-1])
    if voy:
        if units == 0:
            res += "десять"
        elif units == 1:
            res += "одиннадцать"
        elif units == 2:
            res += "двенадцать"
        elif units == 3:
            res += "тринадцать"
        else:
            res += nums[units][:-1] + "надцать"
    elif units != 0:
        res += nums[units]

    return res

for i in range(1001):
    print(get_num_as_str(str(i)))
```

# Работа с функциями

Неизвестное число аргументов

```python
def get_sum(*nums):
    s = 0
    for a in nums:
        s+=a
    return s

get_sum(1,3)
```

```python
def rec(n):
    if n > 0:
        return n + rec(n - 1)
    return 0

rec(20)
```

```python
def print_num_symb(n):
    if n > 0:
        print(n%10)
        print_num_symb(n//10)
        
print_num_symb(25678)
```

```python
def rec_dec_to_bin(n):
    if n/2 != 1:
        return rec_dec_to_bin(n//2) + str(n%2)
    else:
        return "1"
    
rec_dec_to_bin(4)
```
