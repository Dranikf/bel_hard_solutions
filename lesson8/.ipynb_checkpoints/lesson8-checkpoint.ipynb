```python
from math import sin, sqrt, cos, pi

class parallel():
    def __init__(self, side1, side2, angle):
        '''
            Конструктор
        '''
        self.side1 = side1
        self.side2 = side2
        self.angle = angle
    
    def change_side1(self, perc):
        '''
            Изменить первую сторону
        '''
        self.side1 *= perc
    def change_side2(self, perc):
        '''
            Изменитьт вторую сторону
        '''
        self.side2 *= perc
    
    def get_perimetr(self):
        return 2*(self.side1+self.side2)
    
    def get_h(self):
        return sin(self.angle)*self.side2
    
    def get_area(self):
        return self.get_h()*self.side1
        
    def get_diag(self):
        return [
            sqrt(self.side1**2 + self.side2**2 + 2*self.side1*self.side2*cos(self.angle)),
            sqrt(self.side1**2 + self.side2**2 - 2*self.side1*self.side2*cos(self.angle))
        ]
    
par1 = parallel(5,8, pi/60)
par1.change_side1(0.8)
par1.side1
```

```python
class basic_candy():
    '''
        Базовая сладость
    '''
    def __init__(self, candy_name, weight, sugar_part):
        '''
            weigth - вес сладости в гр.
            sugar_part - доля сахара в сладости
        '''
        self.candy_name = candy_name
        self.weight = weight
        self.sugar_part = sugar_part
        
class caramel(basic_candy):
    '''
        Карамелька (по госту весят 10 гр.)
    '''
    def __init__(self, candy_name, taste, sugar_part):
        '''
            taste - вкусс (str)
        '''
        super().__init__(candy_name, 10, sugar_part)
        self.taste = taste
        
class chocolate(basic_candy):
    '''
        Шоколадка
        (по госту содержит 30% сахара)
    '''
    def __init__(self, candy_name, milk_contains, weight):
        super().__init__(candy_name, weight, 0.3)
        self.milk_contains = milk_contains
        
        
class new_year_pack():
    '''
        Пакет сладостей
    '''
    
    def __init__(self, candies):
        '''
            candies - список наследников basic_candy
        '''
        self.candies = candies
        
    def get_weght(self):
        '''
            Суммарный вес конфет без коробки
        '''
        res = 0
        for candy in self.candies:
            res += candy.weight
        return res
    
    def sort_by_sugar(self):
        '''
            Получить сорировку по содержанию сахара
        '''
        surag_counts = [candy.sugar_part for candy in self.candies]
        sorted_keys = sorted(range(len(surag_counts)), key=lambda k: surag_counts[k])
        
        for key in sorted_keys:
            print(self.candies[key].candy_name, self.candies[key].sugar_part)
            
    def get_candy_by_sugar(self, min_sugar, max_sugar):
        '''
            Получисть сладости удостоверяющие заданному
            дипапазону по сахару
        '''
        for candy in self.candies:
            if candy.sugar_part > min_sugar and candy.sugar_part < max_sugar:
                print(candy.candy_name, candy.sugar_part)
    
my_pack = new_year_pack([
    caramel("Красный пищевик", "Малина", 0.9),
    caramel("Сладкий Марио", "Банан", 0.8),
    chocolate("Аленушка", 0.6, 80)
])

print("Вес подарка", my_pack.get_weght(), end = "\n\n")

print("Сортировка по содержанию сахара")
my_pack.sort_by_sugar()

print("\n\n")
print("Диапазон по сахару")
my_pack.get_candy_by_sugar(0.5,0.9)
```

```python
from tkinter import *
from PIL import Image, ImageTk

class basic_candy():
    '''
        Базовая сладость
    '''
    def __init__(self, candy_name, weight, sugar_part, image_file_name):
        '''
            weigth - вес сладости в гр.
            sugar_part - доля сахара в сладости
        '''
        self.candy_name = candy_name
        self.weight = weight
        self.sugar_part = sugar_part
        self.image_file_name = image_file_name



class App:
    
    candy_types = [
        basic_candy("Барбарис", 20, 0.9, "task3_imgs/barbaris.png"),
        basic_candy("Золотой ключик", 10, 0.8, "task3_imgs/gold key.jpg"),
        basic_candy("шоколад Аленушка", 80, 0.5, "task3_imgs/alenka.jpg"),
        basic_candy("шоколад Аленушка2", 80, 0.5, "task3_imgs/alenka.jpg")
    ]
    
    def add_count(self, label, candy_ind):
        
        label['text'] = str(int(label["text"]) + 1)
        self.candyes_counts[candy_ind] += 1
        self.change_photo(self.candy_types[candy_ind].image_file_name)
    
    def add_candy_to_interface(self,ind):
        '''
            Добавить сладость в интерфейс
        '''
        candy_name = self.candy_types[ind].candy_name
        
        cand_fr = Frame()
        lab = Label(cand_fr, text = "0")
        button = Button(
            cand_fr, text = candy_name, 
            command = lambda : self.add_count(lab, ind)
        )
        
        button.pack(side = LEFT, fill = X, expand = 1)
        lab.pack(side = LEFT, fill = X, expand = 1)
        cand_fr.grid(row = ind + 1, column = 0)
    
    def __init__(self):
        self.root = Tk()

        self.frame = Frame(self.root)
        self.frame.grid()
        
        self.candyes_counts = [0]*len(self.candy_types)
        
        self.canvas = Canvas(self.root, height=600, width=700)
        self.change_photo("task3_imgs/barbaris.png")
        
        for i in range(len(self.candy_types)):
            self.add_candy_to_interface(i)
        
        self.root.mainloop()

    def change_photo(self, filename):
        self.image = Image.open(filename)
        self.image.thumbnail([500, 500])
        self.photo = ImageTk.PhotoImage(self.image)
        self.c_image = self.canvas.create_image(0, 0, anchor='nw', image=self.photo)
        self.canvas.grid(row=0, column=0)

app= App()
```
