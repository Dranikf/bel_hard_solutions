```python
import numpy as np
```

`np.empty` - создаст массив без заполнения, т.е. память занимаемая этим массивом не изменяется.

```python
np.empty((3,3))
```

`np.fromfunction` - создаст массив в с использованием функции

```python
np.fromfunction(lambda i, j: i*3 + j, (4,4))
```
