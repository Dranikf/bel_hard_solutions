`set_printoptions`

```python
arr = np.random.rand(1000)

np.set_printoptions(precision = 5)
print("precision = 5")
print(arr)

np.set_printoptions(precision = 2)
print("precision = 2")
print(arr)
```

```python
arr = np.zeros((100,100))
np.set_printoptions(threshold = 1)
print("threshold = 1")
print(arr)
np.set_printoptions(threshold = 10000)
print("threshold = 100")
print(arr)
```

```python
arr = np.random.rand(100)
np.set_printoptions(
    threshold = 1,
    edgeitems = 3
)

print("edgeitems = 3")
print(arr)

np.set_printoptions(
    edgeitems = 5
)
print("edgeitems = 5")
print(arr)
```

```python
arr = np.random.rand(100)
np.set_printoptions(
    threshold = 100,
    linewidth = 50,
    precision = 3
)
print("linewidth = 50")
print(arr)

np.set_printoptions(
    linewidth = 20
)
print("linewidth = 20")
print(arr)
```

```python
arr = np.random.rand(100)/100

np.set_printoptions(
    linewidth = 74,
    suppress = True
)

print("suppress = True")
print(arr)


np.set_printoptions(
    suppress = False
)
print("suppress = False")
print(arr)
```

```python
def test_func(i, j):
    val = i+j
    return np.where(
        val %2 == 0,
        np.NaN, val
    )
    
arr = np.fromfunction(test_func, (4,4))

np.set_printoptions(
    nanstr = "nan"
)
print("nanstr = nan")
print(arr)


np.set_printoptions(
    nanstr = "hello"
)
print("nanstr = hello")
print(arr)
```

```python
np.set_printoptions(
    infstr = "бесконечность"
)
np.array([np.inf])
```

найти самую длинную группу подряд идущих положительных элементов массива;

```python
np.random.seed(10)
arr = np.random.uniform(-10, 10, 20)


cs_res = (arr<0).cumsum()
ind = np.argmax(np.unique(cs_res, return_counts=True)[1])
arr[cs_res == ind][1:]
```
