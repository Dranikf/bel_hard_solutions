Развить приложение созданное в 3-ей практической по tkinter в протейшую среду python разработки.

```python
from tkinter import *
from subprocess import *

def read_file():
    file_name = entry.get()
    fl = open(file_name)
    
    my_text.delete(1.0, END)
    my_text.insert(1.0, fl.read())
    
    fl.close()
    
def write_file():
    file_name = entry.get()
    fl = open(file_name, "w")
    
    fl.write(my_text.get(1.0,END))
    
    fl.close()
    
def run_program():
    file_name = entry.get()
    output=subprocess.check_output(
        "python3 " + file_name, 
        shell=True
    )
    output_text.insert("end", output)

root = Tk()
fr = Frame()
fr.pack(side = TOP)

entry = Entry(fr)
entry.insert(0, "test_files/")
entry.pack(side = LEFT)
Button(
    fr, width = 10, 
    text = "Открыть",
    command = read_file
).pack(side = LEFT)
Button(
    fr, width = 10, 
    text = "Сохранить",
    command = write_file
).pack(side = LEFT)
Button(
    fr, width = 3,
    text = "Run",
    command = run_program
).pack(side = LEFT)

my_text = Text(width = 50)
my_text.pack(side = TOP)

output_text = Text(width = 50, height = 4)
output_text.pack(side = TOP)

root.mainloop()
```

# Практическая работа по `Listbox`

Напишите программу, состоящую из двух списков Listbox. В первом будет, например, перечень товаров, заданный программно. Второй изначально пуст, пусть это будет перечень покупок. При клике на одну кнопку товар должен переходить из одного списка в другой. При клике на вторую кнопку – возвращаться (человек передумал покупать). Предусмотрите возможность множественного выбора элементов списка и их перемещения.

```python
from tkinter import *

def move_elements(_from, to, inds):
    '''
        Перемещение элементов
        по индексам указанным в картеже inds
        из Listbox _from 
        в Listbox to
    '''
    
    for ind in inds:
        to.insert(0,_from.get(ind))
    linds = list(inds)
    linds.sort(reverse=True)
    for ind in linds:
        _from.delete(ind)


def goods_to_purch_click():
    '''
        Обработчик события по
        клику кнопки ">>>"
    '''
    
    move_elements(
        goods_list,
        purch_list,
        goods_list.curselection()
    )
    
def purch_to_goods_click():
    '''
        Обработчик события по
        клику кнопки "<<<"
    '''
    move_elements(
        purch_list,
        goods_list,
        purch_list.curselection()
    )


goods = [
    "apple",
    "bananas",
    "carrot",
    "bread",
    "butter",
    "meat",
    "potato",
    "pineapple",
    "tomato",
    "milk"
]


root = Tk()

goods_list = Listbox(
    selectmode = EXTENDED
)
goods_list.pack(side = LEFT)

for good in goods:
    goods_list.insert(0, good)

buttons_frame = Frame()
buttons_frame.pack(side = LEFT)

purch_list = Listbox(
    selectmode = EXTENDED
)
purch_list.pack(side = LEFT)

Button(
    buttons_frame, 
    text = ">>>",
    command = goods_to_purch_click
).pack()
Button(
    buttons_frame, 
    text = "<<<",
    command = purch_to_goods_click
).pack()

root.mainloop()
```
